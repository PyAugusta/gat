```python
import pandas as pd
from bokeh.layouts import row
from bokeh.models import (
    Range1d,
    WMTSTileSource,
    ColumnDataSource,
    HoverTool,
    ResizeTool
)
from bokeh.plotting import (
    figure,
    curdoc,
    show,
    output_notebook
)
from bokeh.tile_providers import CARTODBPOSITRON
output_notebook()
data = pd.read_csv('gatapp/data/gat_tweets.csv')
data_source = ColumnDataSource(data)
tile_options = {'url': 'https://cartodb-basemaps-b.global.ssl.fastly.net/dark_all/{z}/{x}/{y}.png'}
tile_source = WMTSTileSource(**tile_options)
mercator_extent = dict(start=-20000000, end=20000000, bounds=None)
from mpl_toolkits.basemap import pyproj

wgs84 = pyproj.Proj('+init=EPSG:3395')
```

```python
x = -1.290649610293116*10**7
y = 0.7095660159526624*10**7
x, y
```

```python
test_data = {
    'x': [x], 'y': [y]
}
test_data_source = ColumnDataSource(test_data)
test_data

x_range = Range1d(**mercator_extent)
y_range = Range1d(**mercator_extent)

p = figure(tools='wheel_zoom,pan', x_range=x_range, y_range=y_range)
p.axis.visible = False
#hover_tool = HoverTool(tooltips=[("Tweet" "@text")])
#p.add_tools(hover_tool)
p.add_tile(tile_source)
p.circle(x='x', y='y', size=9, fill_color="#F46B42", line_color="#D2C4C1", line_width=1.5, source=test_data_source)
show(p)
```
